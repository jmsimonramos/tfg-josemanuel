```python
import pandas as pd

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('../data/datasets/rawDataWithConfiguration.csv' , sep=';')
df.head()
```

```python
df.shape
```

```python
# Generamos las variables dummy 
dummy_configuration = pd.get_dummies(df['Configuration'] , prefix='configuration')

df = pd.concat([df , dummy_configuration] , axis=1)
```

```python
dictionary = {"NORTE":0 , "SUR":1}
config_norm = []
for index, row in df.iterrows():
    config_norm.append(dictionary[row['Configuration']])

df["ConfigNORM"] = config_norm 
```

```python
df.head()
```

```python
print(df.isnull().sum())
print(df.shape)
```

```python
sorted_colums = ['operation2','runway','id_leg','operation1','hexid','callsign','type','origin','destination',
    'altitude','ground_speed','vertical_rate',
    'tmp','dew_pt','rel_hum','wind_dir','win_sp','wind_gust','visib','press','sky_lvl','rain','snow','drizzle','fog',
    'time_stamp','part_date_utc','valid_time','time_class','day_week','month','time_ref',
    'id_legNORM','operationNORM','hexidNORM','callsignNORM','typeNORM','originNORM','destinationNORM',
    'operation_Landing','operation_TakeOff',
    'runway_14L','runway_14R','runway_18L','runway_18R','runway_32L','runway_32R','runway_36L','runway_36R','runwayNORM',
    'Configuration','configuration_NORTE','configuration_SUR','ConfigNORM']

df = df[sorted_colums]
df = df.rename(columns={'operation1' : 'operation'})
df = df.drop(columns=["operation2"], axis=1)
```

```python
df.head()
```

```python
print(df.isnull().sum())
print(df.shape)
```

```python
df.to_csv('../data/datasets/cleanDataConfig.csv', sep=';', index=False)
```
