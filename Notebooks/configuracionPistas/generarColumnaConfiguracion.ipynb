```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
df95 = pd.read_csv('../data/resultados/sensibilidad/resume0;95.csv', sep=';')
df95.head()
```

```python
# Generamos la nueva columna obteniendo los datos a partir de la sensibilidad
configuration_colum = []
for index, row in df95.iterrows():
    if row['ConfigNorm'] == 2: # Existe conflicto
        if row['NOperations'] > row['SOperations']:
            configuration_colum.append("NORTE")
        elif row['SOperations'] > row['NOperations']:
            configuration_colum.append("SUR")
        else:
            configuration_colum.append(np.nan)
    else:
        configuration_colum.append(row["Configuracion"])

df95["Configuracion"] = configuration_colum
```

```python
#Comprobamos si tenemos valores desconocidos con el procesamiento anterior
print(df95.isnull().sum())
print(df95.shape)
```

```python
df95['Configuracion'] = df95['Configuracion'].fillna(method='ffill') #Completamos los valores restantes
```

```python
#Comprobamos si tenemos valores desconocidos con el procesamiento anterior
print(df95.isnull().sum())
print(df95.shape)
```

```python
dictionary = {}
for index, row in df95.iterrows(): # Obtenemos la configuracion existente a una determinada fecha en intervalos de 30 mins
    dictionary[row['Day']] = row['Configuracion']
```

```python
# Comprobamos que tenemos todas las fechas y no nos hemos dejado ninguna
print(len(dictionary)) 
print(df95["Day"].unique().shape[0])
```

```python
df = pd.read_csv("../data/datasets/cleanData.csv" , sep=";")
df.head()
```

```python
# Generamos la nueva columna en funcion de la fecha del vuelo y la configuracion en ese momento segun el diccionario anterior
runway_configuration = []
for index, row in df.iterrows():
    runway_configuration.append(dictionary[row['valid_time']])
```
